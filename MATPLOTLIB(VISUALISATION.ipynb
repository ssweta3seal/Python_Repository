```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0.0,5.0)           
x2 = np.linspace(0.0,2.0)

y1 = np.cos(2*np.pi*x1)*np.exp(-x1)      
y2 = np.cos(2*np.pi*x2)

plt.subplot(2,1,1)               ## (rows, columns, panel number)
plt.plot(x1,y1,'o-')
plt.title('Subplot-1')
plt.xlabel('x1')
plt.ylabel('Amp(y1)')

plt.subplot(2,1,2)                       # col=1 ,2nd row 
plt.plot(x2,y2,'.-')
plt.title('Subplot-2')
plt.xlabel('x2')
plt.ylabel('Amp(y2)')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,2*np.pi,0.1)   # start=0, stop=2*pi  , step=0.1
y1 = np.sin(x)
y2 = np.cos(x) 
plt.plot(x,y1,label='Sin') 
plt.plot(x,y2,label='Cos')

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.legend()
plt.title('Sine & Cosine Functions')
plt.show()
```

```python
import matplotlib.pyplot as plt
activites = ['eat','sleep','work','play']
slices = [2,7,11,5]                              #25   # 100%    # 25=100%     #1=4%  # 3=12%      360/100   ,,,    90/20
colors = ['r','g','m','b']    
plt.pie(slices, labels=activites, colors=colors,   startangle=90,    shadow=False,
       explode=(0.7,0,0,0),autopct='%1.2f%%')   
plt.legend()    
plt.show() 
```
